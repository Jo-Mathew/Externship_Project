2. Write a function to Calculate Area and Perimeter of Rectangle

```python
def rect(length, width):
  area = length * width
  perimeter = 2 * (length + width)
  return area, perimeter

length = int(input("Enter the length: "))
width = int(input("Enter the width: "))
area, perimeter = rect(length, width)

print("Area of Rectangle is: ", area)
print("Perimeter of Rectangle is: ", perimeter)
```

3. Write a python Program to find out given No is Prime or Not.

```python
number  = int(input("Enter a number: "))
flag = False

if number == 1:
    print("Number is not prime number")
elif number > 1:
    for i in range(2, number):
        if (number % i) == 0:
            flag = True
            break
    if flag:
        print(number, "is not a prime number")
    else:
        print(number, "is a prime number")
```

prints the largest among them.

5. Write a Python program to calculate electricity bill (accept no of unit from user) according to following criteria: Unit Price
First 100 units  no charge
Next 100 units Rs 5 per Unit
After 200 Units Rs 10 per Unit
(For Example if input unit is 350 then total bill amount is 2000)

```python
# 100: no  (U*0)
# <100: rs.5  (100*0)+(U-100)*5
# <200: rs.10  (100*0)+(100*5)+(u-200)*10
units = int(input("Enter the units: "))
total = 0
if units <= 100:
  print("No Charge")
elif units>100 and units<200:
  total = (units-100)*5
  print("Charge is: ",total)
else:
  total = 500+((units-200)*10)
  print("Charge is: ",total)
```

7. Write a python program to check given word is Palindrome or not

```python
string = input("Enter any word: ")

if string == string[::-1]:
  print("Yes, ", string, " is a palindrome word")
else:
  print("No, ", string, " is a palindrome word")
```

#9. Perform Univariate Analysis

```python
# Distrubution of Houses as per the built year
import matplotlib.pyplot as plt

plt.hist(data['Built Year'], bins=range(1900, 2021, 10), color="Lightblue", edgecolor='black')
plt.xlabel('House Built Year')
plt.ylabel('Count of Houses')
plt.title('Distribution of Built Year of Houses')
plt.grid()
```

We can see that most houses are built around year 2000.

```python
# Distribution of House Prices
import matplotlib.pyplot as plt

plt.hist(data['Price'], bins=30, color="Lightblue", edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of House Prices')
plt.ticklabel_format(style='plain')
plt.xticks(rotation = 50)
```

Most houses price falls in the range of 0 to 1 Million, with some extending to 4M

```python
# Count of Waterfront
import seaborn as sns
waterfront_counts = data['waterfront present'].value_counts()

plt.figure(figsize=(4, 3))
sns.barplot(x=waterfront_counts.index, y=waterfront_counts.values)

plt.title('Number of Waterfront Present')
plt.xlabel('Waterfront Present')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

Most of the houses lacks the facility of waterfront.

```python
# Distribution of Living Area
plt.figure(figsize=(8, 6))
sns.histplot(data=data, x='living area', binwidth=500)
plt.title('Histogram of Living Area')
plt.xlabel('Living Area')
plt.ylabel('Frequency')
plt.show()
```

```python
# Checking the conditions of house
plt.hist(data['condition of the house'])
plt.title("Condition of House")
plt.xlabel("Condition")
plt.ylabel("Count")
```

Most houses in the dataset are in a moderate condition of from 3 to 5.

```python
plt.figure(figsize=(30, 6))
plt.subplot(1,3,1)
sns.countplot(x=data['number of bathrooms'])

plt.subplot(1,3,2)
sns.countplot(x=data['number of bedrooms'])

plt.subplot(1,3,3)
sns.countplot(x=data['number of floors'])
```

##10. Perform Bivariate Analysis

```python
# Living Area VS Price of the house
plt.scatter(data['living area'],data['Price'])
plt.title("Living Area vs Price")
plt.xlabel("Living area")
plt.ylabel("Price")
plt.ticklabel_format(style='plain')
```

We could see a Positive corelation between Living area and Price. As the area of living area increases the Price of the house incrases as well.

```python
sns.scatterplot(x='grade of the house', y='Price', data=data)
plt.ticklabel_format(style='plain')
plt.xlabel('Grade of the house')
plt.ylabel('Price')
plt.title('Scatter Plot for Grade and Price')
plt.show()
```

We could analyze that as the Grade of house is increasing the Price is increasing as well.

```python
sns.lineplot(x = data['condition of the house'],y = data['Price'])
plt.ticklabel_format(style='plain')
plt.title('Line Plot for Condition of house and Price')
```

We could see that as the condition of house increases the price of the house is increasing as well.

```python
# Multi-Variate Analysis between different columns
cols = ['number of bedrooms', 'living area', 'number of views','number of bathrooms', 'waterfront present', 'condition of the house', 'Price']
selected_data = data[cols]
hm = selected_data.corr()
sns.heatmap(hm, annot=True)
```
